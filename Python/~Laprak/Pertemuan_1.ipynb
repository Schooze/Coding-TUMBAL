```python
# variabel global

kata = 'keren'

def myfunc():
    kata = 'fantastic'

    print('Python adalah bahasa yang', kata)


myfunc()

print('Python adalah bahasa yang', kata)
```

```python
# variabel lokal

kata = 'keren'

def myfunc():
    lokal = 'fantastic'

    print('Python adalah bahasa yang', kata)


myfunc()

print('Python adalah bahasa yang', lokal)
```

```python
# Keyword Global

kata = 'keren'

lokal = 0

def myfunc():
    global lokal

    lokal = 'fantastic'
    

    print('Python adalah bahasa yang', lokal)


myfunc()

print('Python adalah bahasa yang', kata)
```

```python
# Bitwise Operator
# AND

a = 13
b = 37

print(format(a, '08b'))
print(format(b, '08b'))
print(format(a & b, '08b'))

print(a & b)
```

```python
# Bitwise Operator
# XOR

a = 13
b = 37

print(format(a, '08b'))
print(format(b, '08b'))
print(format(a ^ b, '08b'))

print(a ^ b)
```

```python
# Bitwise Operator
# NOT

a = 13

print(format(a, '08b'))
print(format(~a, '08b'))

print(~a)
```

```python
# Operator shift left (<<)

a = 13

print(format(a, '08b'))
print(format(a<<2, '08b'))

print(a<<2)
```

```python
# Operator shift right (<<)

a = 13

print(format(a, '08b'))
print(format(a>>2, '08b'))

print(a>>2)
```
